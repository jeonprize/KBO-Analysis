## 삼성 라이온즈의 2020시즌

- 삼성 라이온즈는 2011~2014년 통합 4연패를 ㄷ달성하며 삼성 라이온즈라는 왕조를 열었지만 15년에 발생한 주축 선수 해외원정도박으로 인해 징계와 방출로 15년도 한국 시리즈를 두산에 힘없이 우승을 내줬고 이후 fa로 주축 선수들의 지속적인 유출과 은퇴, 그리고 오랜시간 상위권을 유지 함으로써 신인 선수 지명의 불리함으로 인해 왕조는 끝이나고 성적도 9-9-6-8-8로 떨어졌다
- 그 동안 부진했지만 신인 선수 지명의 우선권이 주어져 좋은 선수들을 키울 수 있었으며 20시즌 데이터를 중시하는 허삼영 감독의 부임과 데이터를 기반으로한 적재제소한 선수 배치를 앞세워 시즌 중반 단독 4위를 달성하며 암흑기 탈출을 꿈꿨으나 체력저하로 인한 타자들의 답답한 공격력으로 인해 결국 8위로 시즌을 마감

- 상승세를 이어가지못하고 8위로 성적이 떨어진 이유는 무엇일까? 생각해보았다.15승을 기록한 외국인 투수, 11승을 달성한 신입급 투수, 한국 최고의 마무리가 지키는 마운드는 기대 이상의 성적을 거두었지만 한 시즌 내내 답답한 모습을 보여준 타자들의 책임이 크다고 생각한다.
- 그래서 삼성 타자들의 데이터를 가지고 삼성 타선의 현실을 보려구 한다. 그리고 FA가 아직 시작하지 않았다고 생각하고 어떤 유형의 선수를 영입해야하는지 분석을 해보겠다.

* 각 팀의 주전임을 알 수 있도록 타석 수를 기준으로 주전을 선발

```python
# 각 포지션 별 리스트 
C_list = ['C']
B1_list = ['1B']
B2_list = ['2B']
B3_list = ['3B']
SS_list = ['SS']
CF_list = ['CF']
LF_list = ['LF']
RF_list = ['RF']
DH_list = ['DH']
```

```python
# 각 포지션 자리 숫자
C_count = 1
B1_count = 1
B2_count = 1
B3_count = 1
SS_count = 1
CF_count = 1
LF_count = 1
RF_count = 1
DH_count = 1

sl_war = []

for index in sl_200.index :
    if sl_200["포지션"][index] in C_list :
        if C_count != 0 :
            sl_war.append(sl_200["WAR"][index])
            C_count -= 1
    elif sl_200["포지션"][index] in B1_list :
            if B1_count != 0 :
                sl_war.append(sl_200["WAR"][index])
                B1_count -= 1
    elif sl_200["포지션"][index] in B2_list :
             if B2_count != 0 :
                sl_war.append(sl_200["WAR"][index])
                B2_count -= 1
    elif sl_200["포지션"][index] in B3_list :
             if B3_count != 0 :
                sl_war.append(sl_200["WAR"][index])
                B3_count -= 1
    elif sl_200["포지션"][index] in SS_list :
             if SS_count != 0 :
                sl_war.append(sl_200["WAR"][index])
                SS_count -= 1
    elif sl_200["포지션"][index] in CF_list :
             if CF_count != 0 :
                sl_war.append(sl_200["WAR"][index])
                CF_count -= 1
    elif sl_200["포지션"][index] in LF_list :
             if LF_count != 0 :
                sl_war.append(sl_200["WAR"][index])
                LF_count -= 1
    elif sl_200["포지션"][index] in RF_list :
             if RF_count != 0 :
                sl_war.append(sl_200["WAR"][index])
                RF_count -= 1
    else :
            if DH_count != 0 :
                sl_war.append(sl_200["WAR"][index])
                DH_count -= 1
    
```
