```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.rc("font",family="Malgun Gothic")
plt.rc("axes",unicode_minus=False)
plt.style.use("fivethirtyeight")
```

```python
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")
```

```python
df = pd.read_csv("ojl_vs_df.csv")
df
```

```python
df = df[['상황', '타석', '타수','타율', '득점', '안타', '2타', '3타', '홈런', '루타', '타점','볼넷','출루','장타','OPS']]
df
```

```python
home_away = df[df["상황"].str.contains('경기')]
home_away
```

```python
plt.style.use('ggplot')

plt.figure(figsize=(14,12))
plt.subplot(331)
plt.title("2020시즌 오재일 홈, 원정 기록 : 타석 수",pad=15)
plt.bar(home_away["상황"],home_away["타석"],color='dodgerblue',width=0.5)

plt.subplot(332)
plt.title("2020시즌 오재일 홈, 원정 기록  : 타율 ",pad=15)
plt.bar(home_away["상황"],home_away["타율"],color='dodgerblue',width=0.5)

plt.subplot(333)
plt.title("2020시즌 오재일 홈, 원정 기록  : 홈런",pad=15)
plt.bar(home_away["상황"],home_away["홈런"],color='dodgerblue',width=0.5)

plt.subplots_adjust(left=0.125, bottom=0.1,  right=0.9, top=0.9, wspace=0.2, hspace=0.4)

plt.subplot(334)
plt.title("2020시즌 오재일 홈, 원정 기록  : 출루율",pad=15)
plt.bar(home_away["상황"],home_away["출루"],color='dodgerblue',width=0.5)

plt.subplot(335)
plt.title("2020시즌 오재일 홈, 원정 기록  : 장타율",pad=15)
plt.bar(home_away["상황"],home_away["장타"],color='dodgerblue',width=0.5)

plt.subplot(336)
plt.title("2020시즌 오재일 홈, 원정 기록  : OPS",pad=15)
plt.bar(home_away["상황"],home_away["OPS"],color='dodgerblue',width=0.5)
```

```python
vs_df = df[df["상황"].str.contains("vs")]
vs_df
```

```python
plt.style.use('ggplot')

plt.figure(figsize=(19,12))
plt.subplot(331)
plt.title("2020시즌 오재일 상대전적 : 타석 수",pad=15)
plt.plot(vs_df["상황"],vs_df["타석"],color='r',marker='o',linewidth=2,markerfacecolor = 'b')

plt.subplot(332)
plt.title("2020시즌 오재일의 상대전적 : 타율 ",pad=15)
plt.plot(vs_df["상황"],vs_df["타율"],color='r',marker='o',linewidth=2,markerfacecolor = 'b')

plt.subplot(333)
plt.title("2020시즌 오재일의 상대전적 : 홈런",pad=15)
plt.plot(vs_df["상황"],vs_df["홈런"],color='r',marker='o',linewidth=2,markerfacecolor = 'b')

plt.subplots_adjust(left=0.125, bottom=0.1,  right=0.9, top=0.9, wspace=0.2, hspace=0.4)

plt.subplot(334)
plt.title("2020시즌 오재일의 상대전적 : 출루율",pad=15)
plt.plot(vs_df["상황"],vs_df["출루"],color='r',marker='o',linewidth=2,markerfacecolor = 'b')

plt.subplot(335)
plt.title("2020시즌 오재일의 상대전적 : 장타율",pad=15)
plt.plot(vs_df["상황"],vs_df["장타"],color='r',marker='o',linewidth=2,markerfacecolor = 'b')

plt.subplot(336)
plt.title("2020시즌 오재일의 상대전적 : OPS",pad=15)
plt.plot(vs_df["상황"],vs_df["OPS"],color='r',marker='o',linewidth=2,markerfacecolor = 'b')
```

```python
delete_df = df[(df["상황"].str.contains("경기"))|(df["상황"].str.contains("vs"))].index
vs_stadium_df = df.drop(delete_df)
vs_stadium_df
```

```python
fig, axes = plt.subplots(5,1,figsize=(18,20))
fig.suptitle('2020시즌 오재일 구장별 타격 기록',fontsize=20)

axes[0].set_title('구장별 타율',pad=15)
g1 = sns.barplot(ax=axes[0],x=vs_stadium_df["상황"], y = vs_stadium_df["타율"])
g1.set(xlabel=None)
g1.set(ylabel=None)

axes[1].set_title('구장별 홈런 수',pad=15)
g2 = sns.barplot(ax=axes[1],x=vs_stadium_df["상황"], y = vs_stadium_df["홈런"])
g2.set(xlabel=None)
g2.set(ylabel=None)

axes[2].set_title('구장별 출루율',pad=15)
g3 = sns.barplot(ax=axes[2],x=vs_stadium_df["상황"], y = vs_stadium_df["출루"])
g3.set(xlabel=None)
g3.set(ylabel=None)

plt.subplots_adjust(left=0.125, bottom=0.1,  right=0.9, top=0.9, wspace=0.2, hspace=0.6)

axes[3].set_title('구장별 장타율',pad=15)
g4 = sns.barplot(ax=axes[3],x=vs_stadium_df["상황"], y = vs_stadium_df["장타"])
g4.set(xlabel=None)
g4.set(ylabel=None)

axes[4].set_title('구장별 OPS',pad=15)
g5 = sns.barplot(ax=axes[4],x=vs_stadium_df["상황"], y = vs_stadium_df["OPS"])
g5.set(xlabel=None)
g5.set(ylabel=None)
```
