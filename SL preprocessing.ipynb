```python
df = pd.read_csv("SL_batter_data.csv")
df.head(30)
```

```python
# import seaborn as sns
```

```python
plt.figure(figsize=(20,9))
sns.barplot(data=df, x="포지션",y="WAR",hue="시즌",ci=None)
```

```python
sns.catplot(data = df, x="포지션",y="WAR",kind = "bar", col="시즌",col_wrap=3,ci=None,sharex=False, sharey=False)
```

```python
reg_df = df[df["타석"] >= 446]
reg_df
```

```python
plt.figure(figsize=(20,9))
sns.barplot(data=reg_df, x="포지션",y="WAR",hue="시즌",ci=None)
plt.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0.,fontsize='x-large')
```

```python
sns.catplot(data = reg_df, x="포지션",y="WAR",kind = "bar", col="시즌",col_wrap=3,ci=None,sharex=False, sharey=False)
```

```python
season15 = reg_df[(reg_df["시즌"] == 15)]
season15
```
